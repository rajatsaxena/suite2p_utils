```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from scipy import stats
```

```python
image1 = io.imread('7-30_000_002.tif')
image2 = np.array(image1)
plt.imshow(image2[2,:,:],cmap='Reds')
mean_image1 = np.mean(image2,axis=0)
image2_deltaF = (image2-mean_image1)/mean_image1
image2_deltaF = (image2_deltaF-image2_deltaF.min())/image2_deltaF.max()
image2_deltaF*=255
image3 = np.concatenate((image1,image2_deltaF),axis=2)
image4 = stats.zscore(image2_deltaF,axis=0)
image5 = np.concatenate((image3,image4),axis=2)
io.imsave('dfF.tif',image5.astype('int8'))
```
